# Exercise 3 – Self Reflection Prompt (Offline)

**Goal:** Show:
- Before summary
- Self critique prompt with explicit criteria
- After improved summary
- Visible improvement notes

**Offline note:** No API calls. We simulate the critique and then provide a revised version.


## Requirements for critique
- Accuracy: no added facts
- Clarity: beginner friendly
- Length: 90 to 120 words
- Format: exactly 2 paragraphs
- Focus: key points only
- Tone: professional and neutral


```python
# ---------------- Before summary ----------------
before_summary = (
    "Capitalism is an economic system where private individuals own businesses and seek profit. "
    "Competition can encourage better products and lower prices. Critics argue it can increase inequality "
    "and allow large companies to gain too much power."
)

print("BEFORE SUMMARY:")
print(before_summary)
print("\nWord count:", len(before_summary.split()))
```

## Self reflection prompt (what I would ask the AI)

"Critique this summary against the requirements, then rewrite it to fully comply.  
First list issues as bullet points. Then output the revised summary as exactly two paragraphs."


```python
# ---------------- Simulated critique (as if from AI) ----------------
critique = [
    "Length is under 90 words (does not meet the 90–120 word requirement).",
    "Not formatted as exactly two paragraphs.",
    "Could be clearer by defining capitalism briefly and balancing pros/cons evenly."
]

print("CRITIQUE:")
for item in critique:
    print("-", item)
```

```python
# ---------------- After summary (revised) ----------------
after_summary = (
    "Capitalism is an economic system in which most businesses are privately owned and prices are largely set "
    "through supply and demand. Supporters argue that competition can encourage innovation, efficiency, and "
    "a wider variety of goods and services.\n\n"
    "Critics argue that capitalism can increase economic inequality and allow large firms to gain influence that "
    "reduces real competition. Because outcomes can vary depending on laws and market conditions, many systems "
    "use regulations and safety nets to limit harms while preserving the benefits of competition."
)

print("AFTER SUMMARY:")
print(after_summary)
print("\nWord count:", len(after_summary.split()))
```

### Demonstrated improvement
- The revised version is within 90–120 words  
- It is exactly 2 paragraphs  
- It keeps a neutral tone and clearly states both supporter and critic viewpoints
